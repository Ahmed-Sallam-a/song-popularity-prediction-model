```python
# build a function to return popular and unpopular songs
def getData(type, data, artists, titles, year=None, date=None, label=None):
    if date==None: #data is a list of dictionary we collocet data into.
        HotChart = billboard.ChartData(type, year=year) #year to use it when collect popular data.
    else:
        HotChart = billboard.ChartData(type, date=date) #date to use it when collect unpopular data.
    for hot in HotChart:
        # To collect the name of artists that exists in the artist's list and choose another song for the same artist.
        # This is to avoid repeat songs and enter artist doesn't exist in artists list.
        if hot.title not in titles and hot.artist in artists: #titles is a list of songs name use it to avoid repeat songs.
                titles.append(hot.title)
                data.append({'Artist': hot.artist, 'Title': hot.title, 'Label': label}) #label is 1 if popular and 0 if unpopular.

    return data, titles
```

```python
from youtube_search import YoutubeSearch

def get_video_link_by_name(video_name):
    results = YoutubeSearch(video_name, max_results=1).to_dict()

    if results:
        video_id = results[0]['id']
        video_link = f"https://www.youtube.com/watch?v={video_id}"
        return video_link

    return None
```

```python
import pandas as pd
import librosa
import numpy as np
import scipy
from tqdm import tqdm
from pychorus import find_and_output_chorus

# Function to download audio from YouTube using the artist and song name
def download_audio(artist, song):
    video_link = get_video_link_by_name(f'{artist} - {song}')
    download_audio_from_youtube(video_link, 'audio')

# Function to find the chorus start time in seconds
def find_chorus_start_time():
    if get_audio_duration('audio.opus')>480:
        raise ValueError("Video length exceeds 8 minutes.")


    #this part of code is flawed and not clean to understand.. it raises an error when chorus duration cant reach 15 sec.. but actually this is required :D
    chorus_start_sec = 15
    while chorus_start_sec >= 0:
        chorus_start_sec = find_and_output_chorus("audio.opus", "output.mp3", chorus_start_sec)
        if chorus_start_sec is not None:
            return chorus_start_sec
        chorus_start_sec -= 1
    return None

# Function to extract soundtrack features from the chorus segment
def extract_soundtrack_features():
    chorus_audio, sr = librosa.load('output.mp3', sr=None)
    audio_features = librosa.feature.mfcc(y=chorus_audio, sr=sr)
    min_val = np.min(audio_features, axis=1)
    mean_val = np.mean(audio_features, axis=1)
    median_val = np.median(audio_features, axis=1)
    max_val = np.max(audio_features, axis=1)
    std_val = np.std(audio_features, axis=1)
    skew_val = scipy.stats.skew(audio_features, axis=1)
    kurtosis_val = scipy.stats.kurtosis(audio_features, axis=1)
    song_track_features = np.concatenate((min_val, mean_val, median_val, max_val, std_val, skew_val, kurtosis_val), axis=0)
    return song_track_features
```

```python
import librosa
import numpy as np
import pandas as pd
from scipy.stats import skew, kurtosis

def extract_audio_features():
    y, sr = librosa.load('output.mp3', sr=None)
    features = {
        'chroma_stft': librosa.feature.chroma_stft(y=y, sr=sr),
        'spectral_centroid': librosa.feature.spectral_centroid(y=y, sr=sr),
        'spectral_bandwidth': librosa.feature.spectral_bandwidth(y=y, sr=sr),
        'rolloff': librosa.feature.spectral_rolloff(y=y, sr=sr),
        'zero_crossing_rate': librosa.feature.zero_crossing_rate(y),
        'chroma_cqt' : librosa.feature.chroma_cqt(y=y, sr=sr),
        'chroma_cens' : librosa.feature.chroma_cens(y=y, sr=sr),
        'mfcc' : librosa.feature.mfcc(y=y, sr=sr),
        'rms' :  librosa.feature.rms(y=y),
        'spectral_contrast':librosa.feature.spectral_contrast(y=y, sr=sr),
        'tonnetz' : librosa.feature.tonnetz(y=y, sr=sr)

    }
    stats = ['min', 'mean', 'median', 'max', 'std', 'skew', 'kurtosis']
    datas= np.empty((0,518), dtype=object)
    for feature_data in features.items():
        for stat in stats:
            if stat == 'min':
                datas=np.append(datas,np.min(feature_data,axis =1),axis=0)
            elif stat == 'mean':
                 datas=np.append(datas,np.mean(feature_data,axis =1),axis=0)
            elif stat == 'median':
                 datas=np.append(datas,np.median(feature_data,axis =1),axis=0)
            elif stat == 'max':
                datas=np.append(datas,np.max(feature_data,axis =1),axis=0)
            elif stat == 'std':
               datas=np.append(datas, np.std(feature_data,axis =1),axis=0)
            elif stat == 'skew':
                datas=np.append(datas,skew(feature_data,axis =1),axis=0)
            elif stat == 'kurtosis':
                datas=np.append(datas, kurtosis(feature_data,axis =1),axis=0)
    return datas
```

```python
# Iterate over the rows of the data
for index, row in tqdm(datatotal.iterrows(), total=len(datatotal)):
    artist = row['Artist']
    song = row['Title']

    try:
        # Download the song
        download_audio(artist, song)
        print(artist+' - '+ song)

        # Find the chorus start time
        chorus_start_sec = find_chorus_start_time()

        if chorus_start_sec is not None:
            # Extract the soundtrack features
            song_track_features = extract_soundtrack_features()

            # Update the dataframe with the extracted features
            for i in range(len(song_track_features)):
                datatotal.loc[index, 'feature'+str(i+1)] = song_track_features[i]
        if index%50 == 0:
            datatotal.to_csv('updated_dataset.csv', index=False)


    except Exception as e:
        print(f"Error processing {artist} - {song}: {str(e)}")
        continue

datatotal.to_csv('dataset1.csv', index=False)
```
